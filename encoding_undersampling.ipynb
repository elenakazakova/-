```python
def df_info(df: pd.DataFrame):
    t = []
    for col in df.columns:
        k_unique = df[col].nunique()
        k_nan = df[col].isna().sum()
        k_zeros = (df[col] == 0).sum()
        d_type = df[col].dtype

        t.append([col, k_unique, k_nan, k_zeros, d_type])

    col_info = pd.DataFrame(t, columns=['columns', 'unique', 'NaN', 'zeros', 'd_type'])
    return col_info

info = df_info(data_median)
info
```

**Target Encoding**

```python
test_data = new_df['TARGET']
test_features = ['TARGET']
train_features = [item for item in new_df.columns if item not in test_features]
train_data = new_df[train_features]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    train_data, test_data, random_state=29
)
```

```python
x_target_train = X_train
x_target_train['TARGET'] = y_train
x_target_train
```

```python
def target_encoding(df, variable, target):
    target_mean = df[target].mean()  # среднее значение целевой переменной по всей выборке
    encoding = df.groupby(variable)[target].mean()  # среднее значение целевой переменной для каждой уникальной категории
    df[variable + "_encoding"] = df[variable].map(encoding)  # создание нового столбца с закодированными значениями
    df[variable + "_encoding"].fillna(target_mean, inplace=True)  # заполнение пропущенных значений средним значением целевой переменной
    return df
```

```python
for col in cat_features:
  variable = col
  target = 'TARGET'
  data_target_train = target_encoding(x_target_train, variable, target)
  #data_target = data_target.dropna(col)
```
